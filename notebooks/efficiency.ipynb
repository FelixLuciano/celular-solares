```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn-v0_8')
```

```python
def get_max(cell: pd.DataFrame):
    return cell.iloc[cell["power"].argmax()]

def plot_max(cell: pd.DataFrame):
    cell_max = get_max(cell)

    plt.subplot(211)
    plt.plot(cell["tension"], cell["current"])
    plt.hlines(cell_max["current"], cell["tension"].min(), cell_max["tension"], label="$i_m$", color="#000", linestyles="dotted")
    plt.vlines(cell_max["tension"], cell["current"].min(), cell_max["current"], label="$U_m$", color="#000", linestyles="dashed")
    plt.scatter([cell_max["tension"]], [cell_max["current"]], c="red")
    plt.grid(True)
    plt.legend()
    plt.xlabel("Tensão [V]")
    plt.ylabel("Corrente [A]")

    plt.subplot(212)
    plt.plot(cell["tension"], cell["power"])
    plt.hlines(cell_max["power"], cell["tension"].min(), cell_max["tension"], label="$P_m$", color="#000", linestyles="dotted")
    plt.vlines(cell_max["tension"], cell["power"].min(), cell_max["power"], label="$U_m$", color="#000", linestyles="dashed")
    plt.scatter([cell_max["tension"]], [cell_max["power"]], c="red")
    plt.grid(True)
    plt.legend()
    plt.xlabel("Tensão [V]")
    plt.ylabel("Potência [W]")
```

```python
cells = pd.read_csv("../data/cells.csv", delimiter=";", decimal=",")

# plt.suptitle("Ensaios", fontsize=14)
plt.scatter(cells["Voc"], cells["Isc"])
plt.grid(True)
plt.xlabel("Tensão [V]")
plt.ylabel("Corrente [A]")
plt.savefig("../docs/public/image/cells.svg")
plt.show()
```

```python
cell_2 = pd.read_csv("../data/cell-2.csv", delimiter=";", decimal=",")
arg_Pmax = cell_2["power"].argmax()
P_max = cell_2.iloc[arg_Pmax]

P_max
```

```python
plt.figure(figsize=(6, 8))
# plt.suptitle("Ensaio com a Célula 2", fontsize=14)
plot_max(cell_2)
plt.savefig("../docs/public/image/cell-2.svg", bbox_inches="tight")
plt.show()
```
